```python
# importing all libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
from sklearn import tree
```

```python
#Reading data from csv file

data = pd.read_csv("Iris.csv")
```

```python
#checking for null values

data.isnull().sum()
```

```python
# showing top 5 data

data.head()
```

```python
# No. of rows and columns

data.shape
```

```python
# Showing only Species data (Dependent Variable)

print(data['Species'])
```

```python
# splitting data in ratio of 75:25 (where 75% data for training while 25% for test)

fc = [x for x in data.columns if x!="Species"]
x= data[fc]
y= data["Species"]
X_train, X_test, Y_train, Y_test = train_test_split(x,y, random_state = 100, test_size = 0.25)
```

```python
# Display of data

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#building Desicion tree model

model1 = DecisionTreeClassifier()
```

```python
# fitting the data

model1.fit(X_train,Y_train)
```

```python
# predicting the value

Y_pred = model1.predict(X_test)
```

```python
# comparing the prediction

data2 = pd.DataFrame({"Actual":Y_test,"Predicted":Y_pred})
data2.head()
```

```python
# Testing the accuracy of model prediction

accuracy_score(Y_test,Y_pred)
```

```python
# Plotting

f_n = ["Sepal length", "Sepal width", "Petal length", "Petal width"]
c_n = ["Setosa", "Versicolor", "Virginica"]
plot_tree(model1,feature_names = f_n, class_names = c_n , filled = True)
```
